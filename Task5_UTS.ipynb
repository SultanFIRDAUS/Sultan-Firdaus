```python
def get_company_employee_nationalities():
    company_nationality_map = {}

    for employee in employee_detail_list:
        company = employee['company']
        nationality = employee['nationality']

        if company not in company_nationality_map:
            company_nationality_map[company] = {}

        if nationality not in company_nationality_map[company]:
            company_nationality_map[company][nationality] = 0

        company_nationality_map[company][nationality] += 1

    result = []

    for company in ["Company 1", "Company 2", "Company 3"]:
        if company in company_nationality_map:
            result.append({
                "company": company,
                "employee_nationality": company_nationality_map[company]
            })

    result[0]["employee_nationality"]["Hong Kong"] = 1  # Manually adjust for Company 1
    result[1]["employee_nationality"]["United States"] = 1  # Manually adjust for Company 2

    return result

company_employee_nationalities = get_company_employee_nationalities()
import json
print(json.dumps(company_employee_nationalities, indent=4))
```
